```python
fruit_data = pd.DataFrame({
    'Fruit': ['Apple', 'Banana', 'Cherry', 'Date', 'Elderberry'],
    'Sales_Q1': [50, 70, 30, 20, 10],
    'Sales_Q2': [60, 65, 35, 25, 15]
})
```

```python
alt.Chart(fruit_data).mark_circle(size=200, color='orange').encode(
    x='Sales_Q1',
    y='Sales_Q2',
    tooltip=['Fruit', 'Sales_Q1', 'Sales_Q2']
).interactive()
```

```python
import altair as alt
import pandas as pd

fruit_data = pd.DataFrame({
    'Fruit': ['Apple', 'Banana', 'Cherry', 'Date', 'Elderberry'],
    'Sales_Q1': [50, 70, 30, 20, 10],
    'Sales_Q2': [60, 65, 35, 25, 15],
    'Region': ['North', 'South', 'East', 'West', 'Central']
})

alt.Chart(fruit_data).mark_circle(size=60).encode(
    x='Sales_Q1',
    y='Sales_Q2',
    color='Region',
    tooltip=['Fruit', 'Sales_Q1', 'Sales_Q2', 'Region']
).interactive()
```
